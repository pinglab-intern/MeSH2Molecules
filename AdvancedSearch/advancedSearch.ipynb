```python
import urllib.request
import urllib
import re
import pandas as pd
import math
import numpy as np
```

```python
CVDfile = './cvd.csv'
df = pd.read_csv(CVDfile)
```

```python
list_ = []
for i in df['name']:
    temp = i.split(',')[0]
    #if len(i.split(',')) > 1: 
    #   temp2 = i.split(',')[1]
    #   if temp2 not in list_:
    #        list_.append(temp2)
    if temp not in list_:
        list_.append(temp)
print(len(list_))
```

```python
for i in list_:
    if i[len(i)-1] == 's':
        list_.append(i[:-1:])
len(list_)
```

```python
searchList = []
for i in list_:
    temp = i.split()
    if len(temp) == 1:
        searchList.append(temp[0])
    else:
        word = temp[0]
        for i in range(1,len(temp)):
            word = word + '+' + 'AND' + '+' + temp[i]
        searchList.append(word)
print(len(searchList))
```

### Pathway

```python
urls_pathway = []
for i in searchList:
    url = "https://reactome.org/content/query?q=" + i + "&species=Homo+sapiens&types=Pathway&cluster=true"
    #print(url)
    urls_pathway.append(url)
```

```python
print(len(urls_pathway))
print(len(list_))
```

```python
dic_pathway = {}
for i in range(len(urls_pathway)):
    url = urls_pathway[i]
    flag = 0
    try:
        request = urllib.request.Request(url)
        response = urllib.request.urlopen(request)
        content = response.read().decode('ISO-8859-1')
        #print(content)
        pattern = re.compile('Filter.*?<strong>(.*?)</strong>.*of.*<strong>(.*?)</strong>')
        pages = re.findall(pattern, content)
        #print(pages)
        print(list_[i],' is successful!')
        
        x = int(pages[0][0])
        n = int(pages[0][1])
        pageSum = math.ceil(n/x)
        
        temp = []

        for j in range(pageSum):
            url_page = url + '&page=' + str(j+1)
            try:
                request = urllib.request.Request(url_page)
                response = urllib.request.urlopen(request)
                content = response.read().decode('ISO-8859-1')
                #print(content)
                pattern = re.compile('<strong>Identifier: </strong>(R-HSA-\d+)')
                items = re.findall(pattern, content)
                for k in items:
                    temp.append(k)

            except urllib.request.URLError as e:
                if hasattr(e,"code"):
                    print(e.code)
                if hasattr(e,"reason"):
                    print(e.reason)
                    
        dic_pathway[list_[i]] = temp

    except urllib.request.URLError as e:
        print(list_[i], 'is failed!')
        #if hasattr(e,"code"):
        #    print(e.code)
        #if hasattr(e,"reason"):
        #    print(e.reason)
```

```python
np.save('dicCVD_pathway.npy',dic_pathway)
```

```python
dicNew = []
c = 0
for i in dic_pathway.keys():
    for j in dic_pathway.keys():
        if (i[:-1:] == j):
            c+=1
            dicNew.append(i)
```

```python
dicNew
```

```python
t = {}
for i in dic_pathway.keys():
    if i in dicNew:
        t[i] = dic_pathway[i]
        for j in dic_pathway[i[:-1:]]:
            if j not in t[i]:
                t[i].append(j)
    else:
        t[i] = dic_pathway[i]
```

```python
s = t.copy()
```

```python
for i in t.keys():
    temp = i+'s'
    if temp in t.keys():
        del s[i]
```
